**SELECTION** **SORT**

```python
def selec_srt(arr):
  n = len(arr)
  for i in range(n):
    min = i
    for j in range(i+1, n):
      if arr[j] < arr[min]:
        min = j
    arr[i], arr[min] = arr[min], arr[i]
  return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print(selec_srt(arr))
```

**INSERTION SORT**

```python
def inser_srt(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i-1
    while j >= 0 and arr[j]>key:
      arr[j+1] = arr[j]
      j -= 1
    arr[j+1] = key
  return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print(inser_srt(arr))
```

**BUBBLE SORT**

```python
def bub_sort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] >arr[j+1]:
        arr[j],arr[j+1] = arr[j+1], arr[j]
  return arr

arr = [69,34,25,12,22,11,90]
print(bub_sort(arr))
```

**QUICK SORT**

```python
def quick_srt(arr, low, high):
  if low < high:
    p = partition(arr, low, high)
    quick_srt(arr, low, p-1)
    quick_srt(arr, p+1, high)
  return arr

def partition(arr,low,high):
  p = arr[low]
  i = low
  j= high
  while True:
    while i <= j and arr[i] <= p:
      i += 1
    while i <= j and arr[j] >= p:
      j -= 1
    if i<=j:
      arr[i], arr[j] = arr[j], arr[i]
    else:
      break
  arr[low], arr[j] = arr[j], arr[low]
  return j


arr = [69,34,25,12,22,11,90]
print(quick_srt(arr, 0, len(arr)-1))
```

**HEAP SORT**

```python
def heapify(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and arr[l] > arr[largest]:
        largest = l
    if r < n and arr[r] > arr[largest]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
def heapSort(arr):

    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0)

def printArray(arr):
    for i in arr:
        print(i, end=" ")
    print()

arr = [9, 4, 3, 8, 10, 2, 5]
heapSort(arr)
print("Sorted array is ")
printArray(arr)
```

**TOPOLOGICAL SORT**

```python
def topological_sort(V,adj,visited,stack):
  visited[V] = True
  for i in adj[V]:
    if not visited[i]:
      topological_sort(i,adj,visited,stack)
  stack.append(V)
  return stack

def topo_sort(V,adj):
  visited = [False]*V
  stack = []
  for i in range(V):
    if not visited[i]:
      topological_sort(i,adj,visited,stack)
  print("Toplogical sorting of the graph is: ")
  while stack:
    print(stack.pop(), end=" ")

V = 6
edges = [[5,2],[5,0],[4,0],[4,1],[2,3],[3,1]]
adj = [[] for i in range(V)]
for i in edges:
  adj[i[0]].append(i[1])
topo_sort(V,adj)
```

**BFS**

```python
def BFS(graph,V,src):
  visited = [False] * V
  queue = []
  queue.append(src)
  visited[src] = True
  path =[]

  while queue :
    u = queue.pop(0)
    # print(u , end =" ")
    path.append(u)
    for v in graph.get(u,[]):
      if visited[v] == False:
        queue.append(v)
        visited[v] = True
  return path

graph = {0:[1,2], 1:[3], 2:[4,5], 3:[], 4:[], 5:[] }
V = 6
src = 0
print(BFS(graph,V,src))
```

**DFS**

```python
def DFS(graph,V,src):
  visited = [False] * V
  stack = []
  stack.append(src)
  path = []

  while stack:
    u = stack.pop()
    # print(u, end=" ")
    path.append(u)
    if visited[u] == False:
      visited[u] = True
    for v in graph.get(u,[]):
      if visited[v] == False:
        stack.append(v)
  return path

graph = {0:[1,2], 1:[3], 2:[4,5], 3:[], 4:[], 5:[] }
V = 6
src = 0
print(DFS(graph,V,src))
```
